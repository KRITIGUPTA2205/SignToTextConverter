```python
import cv2
import mediapipe as mp
import numpy as np
import os
from tqdm import tqdm

DATA_PATH = "ASL_words_dataset"
OUTPUT_PATH = "ASL_processed"
os.makedirs(OUTPUT_PATH, exist_ok=True)

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

SEQUENCE_LENGTH = 30  # frames per sample

def extract_landmarks_from_video(file_path):
    cap = cv2.VideoCapture(file_path)
    sequence = []
    
    with mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.6, min_tracking_confidence=0.6) as hands:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            result = hands.process(frame_rgb)

            if result.multi_hand_landmarks:
                hand = result.multi_hand_landmarks[0]
                landmarks = []
                for lm in hand.landmark:
                    landmarks.extend([lm.x, lm.y, lm.z])
                sequence.append(landmarks)
            else:
                # No hand â†’ append empty frame (zeros)
                sequence.append([0]*63)

    cap.release()

    # Pad/clamp to fixed length
    if len(sequence) < SEQUENCE_LENGTH:
        sequence.extend([[0]*63] * (SEQUENCE_LENGTH - len(sequence)))
    else:
        sequence = sequence[:SEQUENCE_LENGTH]

    return np.array(sequence)

# Process all videos
for word in os.listdir(DATA_PATH):
    word_folder = os.path.join(DATA_PATH, word)
    output_folder = os.path.join(OUTPUT_PATH, word)
    os.makedirs(output_folder, exist_ok=True)

    for file in tqdm(os.listdir(word_folder), desc=f"Processing {word}"):
        if file.endswith(".avi"):
            video_path = os.path.join(word_folder, file)
            data = extract_landmarks_from_video(video_path)

            np.save(os.path.join(output_folder, file.replace(".avi", ".npy")), data)

print("âœ… All video keypoints extracted into", OUTPUT_PATH)
```

```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tqdm import tqdm
```

```python
DATA_PATH = "ASL_processed"
words = sorted(os.listdir(DATA_PATH))
print("Words:", words)

# Load sequences & labels
X, y = [], []

for word in words:
    word_folder = os.path.join(DATA_PATH, word)
    for file in os.listdir(word_folder):
        if file.endswith(".npy"):
            X.append(np.load(os.path.join(word_folder, file)))
            y.append(word)

X = np.array(X)
y = np.array(y)
print("Data shape:", X.shape, "Labels:", y.shape)
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
num_classes = len(le.classes_)
print("Classes:", le.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, shuffle=True, stratify=y_encoded, random_state=42
)
```

```python
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X.shape[1], X.shape[2])),
    Dropout(0.3),

    LSTM(64),
    Dropout(0.3),

    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
es = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=8,
    callbacks=[es],
    verbose=1
)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"âœ… Test Accuracy: {acc*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Extract values
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# Plot Accuracy
plt.figure(figsize=(8,5))
plt.plot(epochs, acc, label="Train Accuracy")
plt.plot(epochs, val_acc, label="Validation Accuracy")
plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Plot Loss
plt.figure(figsize=(8,5))
plt.plot(epochs, loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import cv2
import numpy as np
import mediapipe as mp
from keras.models import load_model

# -------------------- Load LSTM word model --------------------
model = load_model("asl_words_lstm.h5")
word_labels = ['book', 'hello', 'no', 'thankyou', 'yes']

# -------------------- Mediapipe Setup --------------------
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.7)

# -------------------- Sequence buffer --------------------
sequence = []
seq_length = 30

# -------------------- Start Webcam --------------------
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    frame = cv2.flip(frame, 1)
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(img_rgb)

    gesture = "No Gesture"
    hand_found = False
    keypoints = []

    if results.multi_hand_landmarks:
        hand_found = True
        for handLms in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, handLms, mp_hands.HAND_CONNECTIONS)

            for lm in handLms.landmark:
                keypoints.extend([lm.x, lm.y, lm.z])

    # Fill dummy if no hand (so buffer stays consistent)
    if not hand_found:
        keypoints = [0]*63

    # Append frame keypoints
    sequence.append(keypoints)

    # Maintain last 30 frames only
    if len(sequence) > seq_length:
        sequence.pop(0)

    # Predict only when sequence ready
    if len(sequence) == seq_length:
        input_data = np.array(sequence).reshape(1, seq_length, 63)
        pred = model.predict(input_data, verbose=0)
        idx = np.argmax(pred)
        conf = pred[0][idx]

        if np.max(keypoints) == 0:  
            gesture = "No Gesture"
        elif conf > 0.85:
            gesture = word_labels[idx]
        else:
            gesture = "No Match"

    # Display result
    cv2.putText(frame, gesture, (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,255,0), 2)

    cv2.imshow("Word Gesture Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import mediapipe as mp
from keras.models import load_model
from collections import deque

model = load_model("asl_words_lstm.h5")
word_labels = ['book', 'hello', 'no', 'thankyou', 'yes']

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.7)

sequence = deque(maxlen=30)
recent_preds = deque(maxlen=5)
cap = cv2.VideoCapture(0)

MIN_CONF = 0.85

def is_moving(seq):
    arr = np.array(seq)
    return np.mean(np.std(arr, axis=0)) > 0.01  # movement threshold

while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(img_rgb)

    gesture = "No Gesture"
    keypoints = []

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, handLms, mp_hands.HAND_CONNECTIONS)
            for lm in handLms.landmark:
                keypoints.extend([lm.x, lm.y, lm.z])
    else:
        keypoints = [0]*63

    sequence.append(keypoints)

    if len(sequence) == 30 and is_moving(sequence):
        inp = np.array(sequence).reshape(1, 30, 63)
        pred = model.predict(inp, verbose=0)
        idx = np.argmax(pred)
        conf = pred[0][idx]

        if conf > MIN_CONF and max(keypoints) != 0:
            recent_preds.append(word_labels[idx])
        else:
            recent_preds.append("No Match")

        # If last 3 predictions are the same â†’ final prediction
        if len(recent_preds) >= 3 and len(set(list(recent_preds)[-3:])) == 1:
            gesture = list(recent_preds)[-1]
        else:
            gesture = "â€¦"
    else:
        gesture = "No Gesture"

    cv2.putText(frame, gesture, (20,40), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,255,0),2)
    cv2.imshow("Word Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
from tqdm import tqdm

DATA_PATH = "ASL_words_dataset"
OUTPUT_PATH = "ASL_processed"
os.makedirs(OUTPUT_PATH, exist_ok=True)

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

SEQUENCE_LENGTH = 30  # frames per sample

def extract_landmarks_from_video(file_path):
    cap = cv2.VideoCapture(file_path)
    sequence = []
    
    with mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.6, min_tracking_confidence=0.6) as hands:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            result = hands.process(frame_rgb)

            if result.multi_hand_landmarks:
                hand = result.multi_hand_landmarks[0]
                landmarks = []
                for lm in hand.landmark:
                    landmarks.extend([lm.x, lm.y, lm.z])
                sequence.append(landmarks)
            else:
                # No hand â†’ append empty frame (zeros)
                sequence.append([0]*63)

    cap.release()

    # Pad/clamp to fixed length
    if len(sequence) < SEQUENCE_LENGTH:
        sequence.extend([[0]*63] * (SEQUENCE_LENGTH - len(sequence)))
    else:
        sequence = sequence[:SEQUENCE_LENGTH]

    return np.array(sequence)

# Process all videos
for word in os.listdir(DATA_PATH):
    word_folder = os.path.join(DATA_PATH, word)
    output_folder = os.path.join(OUTPUT_PATH, word)
    os.makedirs(output_folder, exist_ok=True)

    for file in tqdm(os.listdir(word_folder), desc=f"Processing {word}"):
        if file.endswith(".avi"):
            video_path = os.path.join(word_folder, file)
            data = extract_landmarks_from_video(video_path)

            np.save(os.path.join(output_folder, file.replace(".avi", ".npy")), data)

print("âœ… All video keypoints extracted into", OUTPUT_PATH)
```

```python
DATA_PATH = "ASL_processed"
words = sorted(os.listdir(DATA_PATH))
print("Words:", words)

# Load sequences & labels
X, y = [], []

for word in words:
    word_folder = os.path.join(DATA_PATH, word)
    for file in os.listdir(word_folder):
        if file.endswith(".npy"):
            X.append(np.load(os.path.join(word_folder, file)))
            y.append(word)

X = np.array(X)
y = np.array(y)
print("Data shape:", X.shape, "Labels:", y.shape)
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
num_classes = len(le.classes_)
print("Classes:", le.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, shuffle=True, stratify=y_encoded, random_state=42
)
```

```python
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X.shape[1], X.shape[2])),
    Dropout(0.3),

    LSTM(64),
    Dropout(0.3),

    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
es = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=8,
    callbacks=[es],
    verbose=1
)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"âœ… Test Accuracy: {acc*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Extract values
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# Plot Accuracy
plt.figure(figsize=(8,5))
plt.plot(epochs, acc, label="Train Accuracy")
plt.plot(epochs, val_acc, label="Validation Accuracy")
plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Plot Loss
plt.figure(figsize=(8,5))
plt.plot(epochs, loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import cv2
import numpy as np
import mediapipe as mp
from keras.models import load_model
from collections import deque

model = load_model("model_lstm.h5")
word_labels = ['book', 'hello', 'no', 'thankyou', 'yes']

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.7)

sequence = deque(maxlen=30)
recent_preds = deque(maxlen=5)
cap = cv2.VideoCapture(0)

MIN_CONF = 0.85

def is_moving(seq):
    arr = np.array(seq)
    return np.mean(np.std(arr, axis=0)) > 0.01  # movement threshold

while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(img_rgb)

    gesture = "No Gesture"
    keypoints = []

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, handLms, mp_hands.HAND_CONNECTIONS)
            for lm in handLms.landmark:
                keypoints.extend([lm.x, lm.y, lm.z])
    else:
        keypoints = [0]*63

    sequence.append(keypoints)

    if len(sequence) == 30 and is_moving(sequence):
        inp = np.array(sequence).reshape(1, 30, 63)
        pred = model.predict(inp, verbose=0)
        idx = np.argmax(pred)
        conf = pred[0][idx]

        if conf > MIN_CONF and max(keypoints) != 0:
            recent_preds.append(word_labels[idx])
        else:
            recent_preds.append("No Match")

        # If last 3 predictions are the same â†’ final prediction
        if len(recent_preds) >= 3 and len(set(list(recent_preds)[-3:])) == 1:
            gesture = list(recent_preds)[-1]
        else:
            gesture = "â€¦"
    else:
        gesture = "No Gesture"

    cv2.putText(frame, gesture, (20,40), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,255,0),2)
    cv2.imshow("Word Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import mediapipe as mp
from tensorflow.keras.models import load_model
from collections import deque

# Load model
model = load_model("model_lstm.h5")

WORDS = ["book", "hello", "no", "thankyou", "yes"]
SEQ_LENGTH = 30
CONF_THRESHOLD = 0.6

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

sequence = deque(maxlen=SEQ_LENGTH)
pred_buffer = deque(maxlen=5)  # smoothing predictions

cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.6, min_tracking_confidence=0.6) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            continue
        
        frame = cv2.flip(frame, 1)
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        res = hands.process(img)

        keypoints = np.zeros(63)
        
        if res.multi_hand_landmarks:
            for hand_landmarks in res.multi_hand_landmarks:
                for i, lm in enumerate(hand_landmarks.landmark):
                    keypoints[i*3]   = lm.x
                    keypoints[i*3+1] = lm.y
                    keypoints[i*3+2] = lm.z
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

        sequence.append(keypoints)

        prediction_text = "Waiting..."

        if len(sequence) == SEQ_LENGTH:
            input_data = np.array(sequence).reshape(1, SEQ_LENGTH, 63)
            pred = model.predict(input_data, verbose=0)[0]
            idx = np.argmax(pred)
            conf = pred[idx]

            if conf > CONF_THRESHOLD:
                pred_buffer.append(idx)
            else:
                pred_buffer.append("none")

            # Smooth prediction
            if pred_buffer.count("none") > 3:
                prediction_text = "No Gesture"
            else:
                most_common = max(set(pred_buffer), key=pred_buffer.count)
                if most_common != "none":
                    prediction_text = WORDS[most_common]

        cv2.putText(frame, prediction_text, (10,40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

        cv2.imshow("ASL Word Recognition", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()
```
